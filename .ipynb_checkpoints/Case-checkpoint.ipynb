```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#os.chdir('C:\Users\c.luo\Desktop\TestCase')
plt.style.use('seaborn-whitegrid')
```

```python
###### Data preparation ######
### Import data
main = pd.read_csv('main.csv', sep = ';')
cust = pd.read_csv('cust.csv', sep = ';')
```

```python
### Restruct cust-file
cust_new = cust.pivot(index = 'cust_id', columns = 'pref')

cust_reset = cust_new.reset_index()
```

```python
### Merge two files
main_merged = pd.merge(main, cust_reset, on = 'cust_id')
```

```python
### Regroup the target
main_merged['target'] = main_merged['boughtItem'].apply(lambda x: 1 if x == '1' else 0)
```

```python
### Make a pivot table
main_merged.pivot_table(index = 'target')
```

```python
###### Viz it! ######
plot_data = main_merged.drop(columns = ['boughtItem', 'cust_id', 'date_lastVisit', 'model1', 'model2', 'boughtItem_03', 'boughtItem_02', 'boughtItem_01'])
```

```python
sns.set(rc={'figure.figsize':(15, 15)}, font_scale = 1.25)

fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for col in range(len(plot_data.columns)):
    if plot_data.columns[col] != 'target':
        fig.add_subplot(5, 6, col+1)
        sns.violinplot(x = 'target', y = plot_data.columns[col], data = plot_data)
    else:
        pass
```

```python
###### Let's do the real thing ######
### Prepare for Scikit 
x_data = main_merged.drop(columns = ['target', 'boughtItem', 'cust_id', 'date_lastVisit', 'model1', 'model2', 'boughtItem_03', 'boughtItem_02', 'boughtItem_01'])
y_data = main_merged['target']
```

```python
### Load some useful stuffs
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
### Split the data for training and testing
sns.set(rc={'figure.figsize':(5, 5)}, font_scale = 1.25)
xtrain, xtest, ytrain, ytest = train_test_split(x_data, y_data, test_size = 0.3, random_state = 0)
```

```python
#### Naive Bayes - Gaussian ####
from sklearn.naive_bayes import GaussianNB
model_0 = GaussianNB()
model_0.fit(xtrain, ytrain)
model_0_pred = model_0.predict(xtest)
print("accurary score is: " + str(accuracy_score(ytest, model_0_pred)))

# Confusion Matrix
mat = confusion_matrix(ytest, model_0_pred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, linewidths=0.5)
plt.title('Gaussian NB', position = (0.5, 1.05))
plt.xlabel('True label', position = (0.5, - 0.05))
plt.ylabel('Predicted label', position = (-0.05, 0.5))
```

```python
#### Naive Bayes - multimonial ####
from sklearn.naive_bayes import MultinomialNB
model_1 = MultinomialNB()
model_1.fit(xtrain, ytrain)
model_1_pred = model_1.predict(xtest)
print(accuracy_score(ytest, model_1_pred))

# Confusion Matrix
mat = confusion_matrix(ytest, model_1_pred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, linewidths=0.5)
plt.title('Multimonial NB', position = (0.5, 1.05))
plt.xlabel('True label', position = (0.5, - 0.05))
plt.ylabel('Predicted label', position = (-0.05, 0.5))
```

```python
#### Logistic Regression ####
from sklearn.linear_model import LogisticRegression
model_2 = LogisticRegression()
model_2.fit(xtrain, ytrain)
model_2_pred = model_2.predict(xtest)
print(accuracy_score(ytest, model_2_pred))

# Confusion Matrix
mat = confusion_matrix(ytest, model_2_pred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, linewidths=0.5)
plt.title('Logistic', position = (0.5, 1.05))
plt.xlabel('True label', position = (0.5, - 0.05))
plt.ylabel('Predicted label', position = (-0.05, 0.5))
```

```python
#### SVM ####
from sklearn.svm import SVC
model_3 = SVC(kernel = 'rbf', C = 1E6)
model_3.fit(xtrain, ytrain)
model_3_pred = model_3.predict(xtest)
print(accuracy_score(ytest, model_3_pred))

# Confusion Matrix
mat = confusion_matrix(ytest, model_3_pred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, linewidths=0.5)
plt.title('SVM', position = (0.5, 1.05))
plt.xlabel('True label', position = (0.5, - 0.05))
plt.ylabel('Predicted label', position = (-0.05, 0.5))
```

```python
#### Random Forest ####
from sklearn.ensemble import RandomForestClassifier

model_4 = RandomForestClassifier(n_estimators=100, random_state=0)
model_4.fit(xtrain, ytrain)
model_4_pred = model_4.predict(xtest)
print(accuracy_score(ytest, model_4_pred))

# Confusion Matrix
mat = confusion_matrix(ytest, model_4_pred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, linewidths=0.5)
plt.title('Random Forest', position = (0.5, 1.05))
plt.xlabel('True label', position = (0.5, - 0.05))
plt.ylabel('Predicted label', position = (-0.05, 0.5))
```
